```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

### Data import and processing

```python
train = pd.read_csv('train.csv', sep=',')
test = pd.read_csv('test.csv', sep=',')
```

```python
train.index = list(train['Id'])
train.index.name = 'ID'
del train['Id']

test.index = list(test['Id'])
test.index.name = 'ID'
del test['Id']
```

```python
for column in list(train.columns)[:-1]:
    if str(train[column].dtypes) == 'int64':
        X = train[column].values.reshape(-1, 1)
        model.fit(X, y)
        scores[column] = [model.score(X, y)]
```

```python
scores = scores.transpose()
scores.columns = ['score']
```

```python
features = list(scores.sort_values('score', ascending = False).head().index)
features
```

we'll use this five features as predictors 

### Predictions

```python
from sklearn.impute import SimpleImputer
from sklearn.model_selection import cross_val_score
```

```python
def predict(X, X_test, y, num):
    
    x_test = SimpleImputer().fit_transform(X_test.values.reshape(-1, 1))  #most features contains 1-2 NaN values in test.csv 
    model.fit(X.values.reshape(-1, 1), y)
    y_predict = model.predict(x_test)
    
    name = 'result' + str(num) + '.csv'
    result = pd.DataFrame(y_predict)
    result.columns = ['SalePrice']
    result.index = X_test.index
    result.to_csv(name, sep=',', 
                  index_label = 'Id', header = True)
    
    return np.mean( cross_val_score(model, X.values.reshape(-1, 1), y, cv=5) )
```

```python
i = 0
CVs = []
for f in features:
    i += 1
    CVs.append(predict(train[f], 
        test[f], 
        train['SalePrice'], i))
```
